### zeub
notebook qui assemble les diffrents csv fait par GTD_preprocessing_zeub en un gros csv final

```python
import pandas as pd

data_folder = './Data/'
GTD1_df = pd.read_csv(data_folder +'2020-12-04_19-29-02GTD.csv')
GTD2_df = pd.read_csv(data_folder +'2020-12-04_19-43-32GTD.csv')

GTD_df = pd.concat([GTD1_df, GTD2_df], axis=1)

columns = GTD_df.columns
columns.values[163]='todel'
GTD_df.columns = columns

GTD_df=GTD_df.drop(labels='todel',axis=1)
GTD_df=GTD_df.set_index('date')
GTD_df.to_csv(data_folder + 'GTD.csv')
GTD_df
```
